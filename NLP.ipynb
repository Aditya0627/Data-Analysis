```python
#from nltk.corpus import stopwords
#from nltk.stem.porter import PorterStemmer
c = []
for i in range(0, 204):
    revie = re.sub('[^a-zA-Z]', ' ', dataset['Complaint'][i])
    revie = revie.lower()
    revie = revie.split() # Splitting the review into a list of words
    ps = PorterStemmer()
    revie = [ps.stem(word) for word in revie if not word in set(stopwords.words('english'))]
    revie = ' '.join(revie) # Joining the list of words
    c.append(revie)
```
